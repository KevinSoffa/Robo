```python
import pyautogui
import time #(Para o programa não dar ERRO)
import pyperclip

pyautogui.PAUSE = 1

# Passo 1:

# Abrir uma nova aba

pyautogui.hotkey('ctrl', 't')

# Entrar no link do sistema

link = 'https://drive.google.com/drive/folders/14oLE59U1RqyRqlBbKpsyymW-mitvbtoh'
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
time.sleep(2)
pyautogui.press('enter')
time.sleep(5)

# Passo 2:

# Clicando na Pasta do arquivo

pyautogui.click(401, 364)
time.sleep(1.3)

# Passo 3:

# Baixando o Arquivo

pyautogui.click(1156, 187) #Clicar nos 3 pontinhos
time.sleep(0.7)
pyautogui.click(1008, 560) #Clicar em fazer dawnload
time.sleep(6)

```

```python
# Passo 4:

# Analisar o Arquivo baixado

import pandas as pd
tabela = pd.read_excel(r'Vendas - Dez.xlsx')
faturamento = tabela ["Valor Final"].sum() # soma da COLUNA 'Valor Final'
quantidade = tabela ["Quantidade"].sum() # soma da COLUNA 'Quantidade'
display(tabela) # Mostrar a tabela
print('===> FATURAMENTO/QUANTIDADE <===')
print('=='*15)
print(f'Faturamento: R${faturamento:,.2f}')
print('=-'*15)
print(f'Quantidade total: {quantidade:,.2f}')
print('=='*15)                            
```

```python
# Passo 5:

import time

# Abrir uma aba

pyautogui.hotkey('ctrl','t')

# Entrar no G-mail

link = 'https://mail.google.com'
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
time.sleep(2)
pyautogui.press ('enter')
time.sleep(5)

# Clicar no botão escrever

pyautogui.click(121, 202 )

# Digitar para quem vai enviar

email_1 = 'kevin@gmail.com'
email_2 = 'fernanda@gmail.com'

pyautogui.write(email_1)
pyautogui.press('tab')
pyautogui.write(email_2)
pyautogui.press('tab')
pyautogui.press('tab')

# Digitar o assunto

assunto = 'RELATÓRIO/ANALISE DE DADOS - Kevin Soffa'
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

# Digitar o corpo do e-mail

texto = f"""Olá, Fernanda.

O faturamento de ontem foi de: R${faturamento:,.2f}
A Quantidade de produtos vendidos foi de: Nº{quantidade:,}

att
Kevin Soffa
"""
pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')

# Clicar em enviar

pyautogui.hotkey('ctrl', 'enter')



#                            =========== FIM DO PROGRAMA  ===========






```
